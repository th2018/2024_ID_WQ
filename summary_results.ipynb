```python
import pandas as pd # read data from a CSV
import matplotlib.pyplot as plt

import os
os.getcwd()
```

```python
data = {
    "site"  :   [""],
"station_nm" : [""],
    "dec_lat_va": [0.0],
    "dec_long_va":  [0.0],
    'max_bias':  [0.0],
        "RMS":  [0.0],
     "rms2021":  [0.0],
        "correlation_coefficient": [0.0]
       }
result = pd.DataFrame(data)
print(result)

rms_data = {"station_nm":[""],
    "2021"  :   [0.0],
   "2021-2023" : [0.0],
       }
rms_df = pd.DataFrame(rms_data)
print(rms_df)
```

```python
#os.chdir('model')
all_files = os.listdir( )
print(all_files)
csv_files = [file for file in all_files if file.endswith(".csv")]
for csv in csv_files:
    df=pd.read_csv(csv)
    result=pd.concat([result,df])
    rms_data0 = {"station_nm":df['station_nm'],
    "2021":df["rms2021"],"2021-2023":df["RMS"],
       }
    rms_df0 = pd.DataFrame(rms_data0)
    rms_df=pd.concat([rms_df0,rms_df])

result=result[ result["dec_lat_va"] >0]
print(result)
print(rms_df)
```

```python
#plt.boxplot(df[['col1', 'col2']], labels=['Column 1', 'Column 2'])
#print(type(rms_df[['2021' ]] ))
rms_df=rms_df[rms_df['2021']>0]

plt.boxplot(rms_df[['2021','2021-2023'] ]    , labels=['2021','2021-2023'] )
plt.ylabel("Root Mean Square Error (deg C)")

#plt.bar( rms_df[['2021' ]])
```
