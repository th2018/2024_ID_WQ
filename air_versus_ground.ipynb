```python
import pandas as pd # read data from a CSV
import matplotlib.pyplot as plt
import dataretrieval.nwis as nwis
```

```python
site  = "13310800"
#Anthropogenic activities have not significantly altered surface and groundwater flows (Kuzis, 1997)

df3 = nwis.get_record(sites=site, service='site')
dec_lat_va=df3['dec_lat_va'][0]
dec_long_va=df3['dec_long_va'][0]
station_nm=df3['station_nm'][0]
print(station_nm)
path=  r"../data_prep/prism_df_summer_" +site +".csv"  # os.path.join( r"../data_prep/prism_df_summer_",  site ,".csv" )
data=pd.read_csv( path )  
print(data.columns)
print(data.tail)
```

```python
data_2023=data[data['waterYear']==2023]
data_2023['Air']=data_2023['tmean..degrees.C.']
data_2023['Stream']=data_2023['X_00010_00003']
data_2023.plot(kind='line', x='Date', y=['Stream', 'Air'])
plt.title(station_nm)
plt.xlabel(" ")
plt.ylabel("Temperature (deg C)")
plt.show()
```
